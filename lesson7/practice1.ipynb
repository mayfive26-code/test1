```python
even_total = 0
odd_total = 0

for i in range(1,101):
    if i % 2 == 0:
        even_total += i
    else:
        odd_total += i
print(f"1加到100偶數的加總為{even_total}")
print(f"1加到100奇數的加總為{odd_total}")
```

```python
multiple_of_threeTotal = 0
notMultiple_of_threeTotal =0
for i in range(1,51):
    if i % 3 == 0:
        multiple_of_threeTotal += i
    else:
        notMultiple_of_threeTotal += i
print(f"1到50中，3的倍數總和為 {multiple_of_threeTotal}")
print(f"1到50中，不是3的倍數總和為 {notMultiple_of_threeTotal}")
```
