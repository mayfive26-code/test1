```python
total:int =0
num:int =0
while True:
    input_value =int(input(f'請輸入第{num+1}個數值:'))
    if input_value < 0 :
        break
    num +=1
    if input_value % 2 ==1:
        continue
    total += input_value
print(f'輸入的次數是{num},，所有輸入的正偶數的加總是:{total}')
    
    
```
